1. Write a Python program that can do the following:
- You have $50
- You buy an item that is $15, that has a 3% tax
- Print how much money you have left, after purchasing the item.

```python
# Money i have
money = 50

# Item price
item_price = 15

# Tax percentage
tax_percent = 3

# Calculate tax amount
tax_amount = item_price * tax_percent / 100

# Total price after tax
total_price = item_price + tax_amount

# Money left
money_left = money - total_price

# Print result
print("Tax amount:", tax_amount)
print("Total price after tax:", total_price)
print("Money left:", money_left)
```

2. Ask the user how many days until their birthday
- Print an approx. number of weeks until their birthday
- 1 week is = to 7 days.

```python
# Ask user input
days = int(input("Enter number of days until your birthday: "))

# Convert days into weeks
weeks = days / 7

# Print result
print("Approximate weeks until your birthday:", weeks)
```

Grades:
- A = 90 - 100

```python
# Create grade variable
grade = 85

# Check grade
if grade >= 90 and grade <= 100:
    print("Letter Grade: A")
elif grade >= 80 and grade <= 89:
    print("Letter Grade: B")
elif grade >= 70 and grade <= 79:
    print("Letter Grade: C")
elif grade >= 60 and grade <= 69:
    print("Letter Grade: D")
elif grade >= 0 and grade <= 59:
    print("Letter Grade: F")
else:
    print("Invalid grade")
```

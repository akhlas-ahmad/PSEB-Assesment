Q4. Calculate the Sum of Digits of a Three-Digit Number


```python
# input from the user
number = int(input("Enter a three numbers: "))

# Extract digits using arithmetic
digit1 = number // 100
digit2 = (number // 10) % 10
digit3 = number % 10

# sum
sum_of_digits = digit1 + digit2 + digit3

print("Sum of digits:", sum_of_digits)
```

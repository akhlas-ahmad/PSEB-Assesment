Part 1: Working with Variables, Input, and Strings

```python
# 1. Prompt for user input 
# Ask user to enter full name
full_name = input("Enter your full name: ")


# Print greeting
print("Hello,", full_name + "! Welcome to the Python assignment.")
```

```python
# 2. String methods
print("Your name in uppercase:", full_name.upper())
print("Your name in lowercase:", full_name.lower())
print("Length of your name:", len(full_name))
print("Name with hyphens:", full_name.replace(" ", "-"))
```

```python
#3. Type casting and numeric operations
birth_year = input("Enter your birth year: ")
birth_year = int(birth_year)

current_year = 2026
age = current_year - birth_year

print("Current year is", current_year, ", so you are approximately", age, "years old.")
```

Part 2: Lists and Indexing

```python
#1.	Create a list from user input

fruits_input = input("Enter three of your favorite fruits: ")
fruit_list = fruits_input.split(" ")

print("Your list of fruits:", fruit_list)
```

```python
#2.	List indexing and slicing

print("First fruit:", fruit_list[0])
print("Last fruit:", fruit_list[-1])
print("Second fruit:", fruit_list[1])
print("First two fruits:", fruit_list[0:2])
```

```python
#3.	List methods

print("Adding a new fruit...")
fruit_list.append("strawberry")
print("Updated list:", fruit_list)

print("Removing one fruit...")
fruit_list.remove(fruit_list[1])
print("Updated list:", fruit_list)

print("Length of the list is now", len(fruit_list))
```

Part 3 (Optional Challenge): Combining Concepts

```python
# Letters List

print("Creating a letters list from your full name...")

letters_list = list(full_name.replace("", ""))

print("First 3 letters:", letters_list[:3])
print("Last 3 letters:", letters_list[-3:])

#Modify & Reconstruct

for i in range(len(letters_list)):
    if i < 3:
        letters_list[i] = letters_list[i].upper()
    else:
        letters_list[i] = letters_list[i].lower()

modified_name = "".join(letters_list)

print("Modified name:", modified_name)

#Numbers, Sum & Average

numbers_input = input("Enter 5 numbers (space-separated): ")
numbers_list = numbers_input.split()

for i in range(len(numbers_list)):
    numbers_list[i] = int(numbers_list[i])

total = sum(numbers_list)
average = total / len(numbers_list)

print("Sum of your numbers:", total)
print("Average of your numbers:", average)
```

5.	List creation, indexing, and methods 

```python
# Creating a list
numbers = [10, 20, 30, 40]

# Indexing
print('First element:', numbers[0])
print('Last element:', numbers[-1])

# Slicing
print('First two elements:', numbers[:2])

# Append
numbers.append(50)
print('After append:', numbers)

# Remove
numbers.remove(20)
print('After remove:', numbers)

# Length
print('Length of list:', len(numbers))
```

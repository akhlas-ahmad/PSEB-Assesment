```python
# Function to calculate grade
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"

marks = int(input("Enter student marks: "))

grade = calculate_grade(marks)

print("Student Grade is:", grade)
```

```python
# Function to check even or odd
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False

num = int(input("Enter a number: "))

result = is_even(num)

print("Is the number even?", result)
```

```python
# Function to calculate factorial
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result

num = int(input("Enter a number: "))

answer = factorial(num)

print("Factorial is:", answer)
```

```python
# Function to find maximum number
def find_max(a, b, c):
    if a >= b and a >= c:
        return a  
    elif b >= a and b >= c:
        return b  
    else:
        return c

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

maximum = find_max(num1, num2, num3)

print("Largest number is:", maximum)
```

```python
# Function to count words
def count_words(sentence):  
    words = sentence.split()   
    return len(words)        

text = input("Enter a sentence: ")

total_words = count_words(text)

print("Total words:", total_words)
```

```python
# Function to check prime number
def is_prime(n):  
    if n <= 1:
        return False  
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

result = is_prime(num)

print("Is the number prime?", result)
```

```python
def list_sum(numbers):
    total = 0
    for num in numbers:
        total = total + num
    return total

user_input = input("Enter numbers: ")

numbers = list(map(int, user_input.split()))

print("Sum of list is:", list_sum(numbers))
```

```python
# Function to check password strength
def check_password(password):
    if len(password) < 8:
        return "Weak"
    digit_count = 0
    special_count = 0
    upper_count = 0
    
    special_characters = "!@#$%^&*()_+-=[]{}|;:,.<>?/"
    
    for char in password:   
        if char.isdigit():
            digit_count += 1
        elif char.isupper():
            upper_count += 1
        elif char in special_characters:
            special_count += 1
    if digit_count >= 1 and special_count >= 2 and upper_count >= 1:
        return "Strong"
    else:
        return "Weak"

pwd = input("Enter password: ")

result = check_password(pwd)

print("Password Strength:", result)
```

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    f = (c * 9/5) + 32
    return f

celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = celsius_to_fahrenheit(celsius)

print("Temperature in Fahrenheit:", fahrenheit)
```

```python
def remove_duplicates(data):
    new_list = []
    for item in data:
        if item not in new_list:
            new_list.append(item)
    return new_list

user_input = input("Enter numbers: ")

data_list = list(map(int, user_input.split()))

result = remove_duplicates(data_list)

print("List without duplicates:", result)
```
